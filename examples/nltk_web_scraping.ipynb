```python
import urllib.request#se aplica para paginas html est√°ticas, no funciona sobre CMS's
```

```python
response=urllib.request.urlopen('http://php.net/')
#response=urllib.request.urlopen('https://www.epn.edu.ec')
```

```python
html=response.read()
```

```python
print(html)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(html,"html5lib")
```

```python
text=soup.get_text(strip=True)
```

```python
print(text)
```

```python
tokens=[t for t in text.split()]#desde aqu√≠ empieza el trabajo de nltk
```

```python
tokens
```

```python
freq=nltk.FreqDist(tokens)
```

```python
for key,val in freq.items():
  print(str(key)+':'+str(val))
```

```python
freq.plot(20,cumulative=False)
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download("stopwords")
```

```python
stopwords.words('english')
```

```python
clean_tokens = tokens[:]
for token in tokens:
    if token in stopwords.words('english'):
        clean_tokens.remove(token)
```

```python
freq = nltk.FreqDist(clean_tokens)
for key,val in freq.items():
    print (str(key) + ':' + str(val))
```

```python
freq.plot(20,cumulative=False)
```
