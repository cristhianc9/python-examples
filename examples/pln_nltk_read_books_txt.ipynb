```python
palabras=miCorpus.words()#extraido palabra por palabra
```

```python
palabras=[word for word in palabras if len(word)>4]
```

```python
fdist=nltk.FreqDist(palabras)
```

```python
print(fdist)
```

```python
for word, frequency in fdist.most_common(20):
    print(('%s--->%d' % (word, frequency)).encode('latin-1'))
```
